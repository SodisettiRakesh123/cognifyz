```python
import pandas as pd

checkins_df = pd.read_csv('/content/checkins.csv')
ratings_df = pd.read_csv('/content/ratings.csv')
users_df = pd.read_csv('/content/users.csv')
venues_df = pd.read_csv('/content/venues.csv')

print("checkins_df:")
display(checkins_df.head())

print("\nratings_df:")
display(ratings_df.head())

print("\nusers_df:")
display(users_df.head())

print("\nvenues_df:")
display(venues_df.head())
```

```python
print("checkins_df info:")
checkins_df.info()

print("\nratings_df info:")
ratings_df.info()

print("\nusers_df info:")
users_df.info()

print("\nvenues_df info:")
venues_df.info()
```

```python
# Merge ratings with users and venues dataframes
merged_df = pd.merge(ratings_df, users_df, left_on='user_id', right_on='id', how='left')
merged_df = pd.merge(merged_df, venues_df, left_on='venue_id', right_on='id', how='left')

# Drop redundant id columns
merged_df.drop(['id_x', 'id_y'], axis=1, inplace=True)

print("Merged DataFrame with user and venue information:")
display(merged_df.head())

print("\nMissing values after merging:")
print(merged_df.isnull().sum())
```

```python
from scipy.sparse import csr_matrix

# Create a sparse user-item matrix
# We need to map user_id and venue_id to unique integer indices
merged_df['user_index'] = merged_df['user_id'].astype('category').cat.codes
merged_df['venue_index'] = merged_df['venue_id'].astype('category').cat.codes

user_item_sparse_matrix = csr_matrix((merged_df['rating'], (merged_df['user_index'], merged_df['venue_index'])))

print("Sparse User-Item Matrix:")
print(user_item_sparse_matrix)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

# Sample text data (replace with actual venue descriptions if available)
sample_text_data = [
    "This is a great restaurant with delicious food.",
    "I enjoyed the atmosphere and the service.",
    "The food was amazing and the staff was friendly."
]

# Create a TF-IDF vectorizer
tfidf_vectorizer = TfidfVectorizer()

# Fit and transform the text data
tfidf_matrix = tfidf_vectorizer.fit_transform(sample_text_data)

print("TF-IDF Matrix:")
print(tfidf_matrix)

print("\nFeatures (words):")
print(tfidf_vectorizer.get_feature_names_out())
```

```python
from sklearn.decomposition import TruncatedSVD
import numpy as np

# We'll use TruncatedSVD for matrix factorization on the sparse matrix
# n_components is the number of latent factors. You can experiment with this value.
svd = TruncatedSVD(n_components=50, random_state=42)
svd.fit(user_item_sparse_matrix)

# Function to get recommendations for a user
def recommend_venues_svd(user_id, n_recommendations=10):
    # Get the user's index from the original merged_df
    if user_id not in merged_df['user_id'].unique():
        return "User ID not found."

    user_index = merged_df[merged_df['user_id'] == user_id]['user_index'].iloc[0]

    # Get the user's latent factors
    user_factors = svd.transform(user_item_sparse_matrix[user_index, :])

    # Predict ratings for all venues
    predicted_ratings = user_factors @ svd.components_

    # Get venue IDs and indices
    venue_indices = merged_df['venue_index'].unique()
    venue_ids = merged_df.drop_duplicates('venue_id').set_index('venue_index').loc[venue_indices]['venue_id']

    # Create a pandas Series for easier manipulation
    predicted_ratings_series = pd.Series(predicted_ratings[0], index=venue_ids.values)


    # Get venues the user has already rated
    rated_venues = merged_df[merged_df['user_id'] == user_id]['venue_id'].tolist()

    # Filter out venues the user has already rated
    unrated_venues_predicted_ratings = predicted_ratings_series.drop(rated_venues, errors='ignore')

    # Recommend top n venues with the highest predicted ratings
    recommended_venues = unrated_venues_predicted_ratings.nlargest(n_recommendations)

    return recommended_venues

# Example usage: Get recommendations for a sample user (replace with an actual user_id from the data)
sample_user_id = merged_df['user_id'].iloc[0] # Using the first user in the merged_df as an example
recommendations = recommend_venues_svd(sample_user_id)

print(f"\nRecommendations for user {sample_user_id}:")
print(recommendations)
```

```python
# Get recommendations for the first 5 unique users
unique_user_ids = merged_df['user_id'].unique()
num_users_to_recommend = 5

for user_id in unique_user_ids[:num_users_to_recommend]:
    recommendations = recommend_venues_svd(user_id)
    print(f"\nRecommendations for user {user_id}:")
    print(recommendations)
```
